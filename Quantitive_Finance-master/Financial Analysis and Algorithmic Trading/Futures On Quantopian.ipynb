# Continous Futures

```python
from quantopian.research.experimental import continuous_future
```

```python
help(continuous_future)
```

```python
print(continuous_future.__doc__)
```

```python
continuous_ng = continuous_future('NG',offset=0,roll='volume',adjustment='mul')
continuous_ng
```

```python
ng_cont_active = history(continuous_ng,
                         fields=['contract','price','volume'],
                         frequency='daily',
                         start='2016-10-01',
                         end='2017-08-01'
                        )
```

```python
ng_cont_active.head()
```

```python
ng_cont_active['contract'].unique()
```

```python
ng_cont_active['price'].plot()
```

```python
ng_cont_active['volume'].plot()
```

```python
ax = ng_cont_active['volume'].plot(ls='--',c='black',lw=3)
ng_consecutive_contract_volume.plot(ax=ax,xlim=('2016-10-01','2017-09-01'))
```

```python
ng_consecutive_contract_price = history(ng_contracts,
                                        fields='price',
                                        frequency='daily',
                                        start='2016-01-01',
                                        end='2017-08-01')
```

```python
ax = ng_cont_active['price'].plot(ls='--',c='black',lw=3)
ng_consecutive_contract_price.plot(ax=ax,xlim=('2016-10-01','2017-09-01'));
```

```python
continous_ng_unadjusted = continuous_future('NG',offset=0,roll='volume',adjustment=None)
```

```python
ng_unadjusted_history = history(continous_ng_unadjusted,
                                        fields=['contract','price'],
                                        frequency='daily',
                                        start='2016-01-01',
                                        end='2017-08-01')
```

```python
ng_unadjusted_history.head()
```

```python
ng_unadjusted_history['price'].plot();
```

```python
pivot_unadj = ng_unadjusted_history.pivot(index=ng_unadjusted_history.index,columns='contract')
```

```python
ax = pivot_unadj.plot();
ng_unadjusted_history.plot(ax=ax,ls='--',c='black',figsize=(15,10));
```
