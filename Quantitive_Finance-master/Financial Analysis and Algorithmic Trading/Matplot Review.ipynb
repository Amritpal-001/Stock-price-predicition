```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2 
```

```python
x,y
```

```python
# Funcational Approach
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()
# left bottom width height
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_title("title")
axis.set_xlabel('X label')
axis.set_ylabel("Y label")
```

```python
fig = plt.figure()
# left bottom width height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger plot")
axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
#Tuple unpacking
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
plt.tight_layout()
```

## Figure Size & DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('data/my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x, x**3, label='X Cubed')

# add legend
#check docs for what 1- 10 locations mean
ax.legend(loc=2)
```

## Plot Apperance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',linewidth=8,alpha=0.7,linestyle='--',marker='o',markersize=20,
       markerfacecolor='black',markeredgewidth=3,markeredgecolor='black')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.plot(x,y,color='green',linewidth=1,linestyle='--',marker='+',markersize=2,
       markerfacecolor='black',markeredgewidth=3,markeredgecolor='black')
```
